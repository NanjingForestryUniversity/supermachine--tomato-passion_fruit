```python
import numpy as np
import cv2
import copy
import pandas as pd
import os
```

```python

# 指定文件夹路径
folder_path = 'data/new/green'  
mask_folder_path = 'mask/new/green'  
output_path = 'result/new/green/green.xlsx'

# 获取文件夹中的所有文件
files = os.listdir(folder_path)
files1 = os.listdir(mask_folder_path)

number = 104

D =[]
for i in range(number):
    with open(folder_path + '/' + str(i+1) + '.hdr', 'r') as hdr_file:
        lines = hdr_file.readlines()
        for line in lines:
            if line.startswith('lines'):
                height = int(line.split()[-1])
            elif line.startswith('samples'):
                width = int(line.split()[-1])
            elif line.startswith('bands'):
                bands = int(line.split()[-1])
        raw_image = np.fromfile(folder_path + '/' + str(i+1) + '.raw', dtype='uint16')
        mask = cv2.imread(mask_folder_path + '/' + 'mask_'+ str(i+1) + '.png', cv2.IMREAD_GRAYSCALE)
        formatImage = np.zeros((height, width, bands))
        for row in range(0, height):
            for dim in range(0, bands):
                formatImage[row,:,dim] = raw_image[(dim + row*bands) * width:(dim + 1 + row*bands)* width]
        # 创建一个和光谱图像形状相同的全零数组，用于存储目标区域的光谱图像
        spectrum_image = np.zeros_like(formatImage)
        # 将掩码图像扩展到和光谱图像形状相同，以便进行按位与操作
        mask = np.stack([mask]*bands, axis=2)
        # 使用掩码提取目标区域的光谱图像
        spectrum_image = np.where(mask == 255, formatImage, 0)
        # print(tomato_spectrum_image.shape)
        # cv2.imshow('', tomato_spectrum_image[:, :, 145])
        # cv2.waitKey(0)
        data_save = []
        for j in range(bands):
            data = copy.deepcopy(spectrum_image[:, :, j])
            data[data>0] = 1
            num = np.sum(data)
            print(num)
            average_values_tomato = np.sum(spectrum_image[:, :, j])/num
            print(str(i)+"目标区域"+"波段"+str(j+1)+"的平均值:", average_values_tomato)
            data_save.append(average_values_tomato)
        print(data_save)
        D.append(data_save)
# 创建一个DataFrame来存储结果
# D_array = np.array(D).reshape(224, 120)
df = pd.DataFrame(D)
#转置
df = df.transpose()

# 将DataFrame导出为Excel文件
df.to_excel(output_path, index=False)
```

```python
## 数据拼接
# 指定Excel文件的文件夹路径
folder_path = 'average_result1124'  # 替换成你的文件夹路径
# 获取文件夹下所有Excel文件的文件名
excel_files = [file for file in os.listdir(folder_path) if file.endswith('.xlsx') or file.endswith('.xls')]


# 创建一个空列表，用于存储每个Excel文件的数据框
dfs = []
for i in range(1,40):
    df = pd.read_excel(folder_path + '/'+str(i) +'.xlsx')
    # 将数据框添加到列表中
    dfs.append(df)

# 使用pandas.concat函数将列表中的所有数据框按列合并
result = pd.concat(dfs, axis=1)

# 将结果保存为新的Excel文件
result.to_excel('merged1124.xlsx', index=False)
```
